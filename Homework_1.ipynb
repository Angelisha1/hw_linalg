```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as la 
```

```python
x = np.array([1,1])
y = np.array([2,0])
z = np.array([0,2])

plt.figure()

plt.plot(x[0], x[1], 'o-r', label='x')
plt.plot(y[0], y[1], '.-b', label='y')
plt.plot(z[0], z[1], 'x-g', label='z')
plt.legend()
plt.grid()
```

```python
s = x + y + z

plt.figure()

plt.plot(s[0], s[1], 'o-m', label='s')
plt.legend()
plt.grid()
```

```python
angle_xy = np.rad2deg(np.arccos(np.dot(x,y) / (la.norm(x) * la.norm(y))))
angle_xy
```
